```python
import numpy as np
```

```python
a = np.array([[9, 6, 0, 0, 0],
              [0, 9, 0, 0, 0],
              [0, 0, 9, 0, 0],
              [0, 0, 0, 9, -5],
              [0, 0, 0, 0, 9]])
b = np.array([[52, 4, 6, 8, 9, 12],
              [2, 54, 6, 8, 9, 12],
              [2, 4, 56, 8, 9, 12],
              [2, 4, 6, 58, 9, 12],
              [2, 4, 6, 8, 59, 12],
              [2, 4, 6, 8, 9, 62]])
```

```python
# compute eigensystem
[values, vectors] = np.linalg.eig(a)

print("The eigenvalues are: ", values.round(3))
print("The eigenvectors are: \n", vectors.round(3))

print("Verification: ")
for i in range(len(a)):
    ax = np.dot(a.copy(), vectors[:,i])
    lambdax = np.dot(values[i], vectors[:,i])
    print((ax - lambdax).round(1))
```

```python
[values, vectors] = np.linalg.eig(b)

print("The eigenvalues are: ", values.round(3))
print("The eigenvectors are: \n", vectors.round(3))
newVector = vectors[:,2] + vectors[:,3]

print("Verification: ")
for i in range(len(b)):
    bx = np.dot(b.copy(), vectors[:,i])
    lambdax = np.dot(values[i], vectors[:,i])
    print((bx - lambdax).round(1))

print("New Vector Verification: \n", (np.dot(b.copy(), newVector) - np.dot(values[2], newVector)).round())
```
